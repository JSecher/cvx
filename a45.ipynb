```python
import numpy as np
import matplotlib.pyplot as plt
import utils

utils.create_output_dir("a45")
```

```python
def gi_diff(lam, alpha):
    val = None
    if alpha > lam:
        val = alpha - lam
    elif alpha < -lam:
        val = alpha + lam
    else:
        val = 0
    return val - 1

def g(lam, alpha):
    val = None
    if alpha > lam:
        val = -1/2 * lam**2 + lam * alpha
    elif alpha < -lam:
        val = -1/2 * lam**2 - lam * alpha
    else:
        val = 1/2 * alpha**2
    return val - lam


```

```python
x = np.linspace(0, 5, 100)
alphas = [-1, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0]
for alpha in alphas:
    y = np.array([gi_diff(i, alpha) for i in x])
    plt.plot(x, y, label=f"alpha={alpha}")
plt.legend()
plt.ylabel("$g_i(\lambda)$")
plt.xlabel("$\lambda$")

filename = utils.get_file_path("a45_g_diff.pdf")
plt.savefig(filename, bbox_inches="tight")
plt.show()
```

```python
x = np.linspace(0, 2, 100)
alphas = [-1, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0]
for alpha in alphas:
    y = np.array([g(i, alpha) for i in x])
    plt.plot(x, y, label=f"alpha={alpha}")
plt.legend()
plt.ylabel("$g(\lambda)$")
plt.xlabel("$\lambda$")

filename = utils.get_file_path("a45_g.pdf")
plt.savefig(filename, bbox_inches="tight")
plt.show()
```
